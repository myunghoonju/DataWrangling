```python
#None is a common default value for function args

def add_and_maybe_multiply(a,b,c=None):
  result = a + b
  if c is not None:
    result = result * c

  return result
```

```python
add_and_maybe_multiply(1,2)
```

```python
#Binary search and maintaining a sorted list
import bisect

c = [1,2,2,2,3,4,7]
bisect.bisect(c,2)
print(bisect.bisect(c,8)) #where an element should be inserted to keep it sorted
```
